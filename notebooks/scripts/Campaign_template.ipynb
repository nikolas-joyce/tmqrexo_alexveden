```python
swarms = {}
for file in os.listdir('./swarms/'):
    if '.swm' not in file:
        continue
    swarm_name = file.replace('.swm','')
    swarms[swarm_name] = SwarmManager.load(os.path.join('swarms', file))
```

```python
passive_exo = {}

exos_mat_files = [
    '../../mat/strategy_270225.mat',
    #'../mat/strategy_2060248.mat',    
]

print('Passive EXO list')
for matfile in exos_mat_files:    
    exo = EXOInfo.from_matfile(matfile)
    passive_exo[exo.exo_name()] = exo.exo_price_index()
    
    print(exo.exo_name())
```

```python
exo.data.fut_price.plot()
```

```python

campaign_exposure = {
     'EU6_BearishCollarBrokenWing_SwingPoint_Long_bearish_failure_Aggresive_Long (Small SPL)': 1,
     'EU6_BearishCollarBrokenWing_SwingPoint_Short_bearish_breakout_Aggresive_Short (large SPh)': 1,
     #'EP_BearishCollarBrokenWing_SwingPoint_Short': 1,
     
     #
     # Passive EXO positions
     #
     #'EP_BearishCollarBrokenWing_EXO': -1, # Short 1 EXO units
     
}
```

```python
#def campaign_equity(swarms, exposure):

campaign_dict = {}
campaign_stats = {'NetProfit': 0.0, 'TradesCount': 0, 'CommissionSum': 0.0}

for camp_name, exposure in campaign_exposure.items():
    if camp_name in swarms:
        swarm_name = camp_name
        smgr = swarms[swarm_name]
        campaign_dict[swarm_name] = smgr.swarm_picked.diff().sum(axis=1) * exposure
        s = smgr.get_swarm_stats(smgr.swarm_picked_stats)
        
        campaign_stats['NetProfit'] += s['NetProfit']
        campaign_stats['TradesCount'] += s['TradesCount']
        campaign_stats['CommissionSum'] += s['CommissionSum']
        
    elif camp_name in passive_exo:
        exo_equity = passive_exo[camp_name]
        campaign_dict[camp_name] = exo_equity.diff() * exposure
    
    
campaign_portfolio = pd.DataFrame(campaign_dict).cumsum()
campaign_equity = campaign_portfolio.sum(axis=1)
```

# Global filter based on campaign equity

```python
GLOBAL_FILTER_CONTEXT = {
'global_filter_function': SwarmFilter.volatility_chandelier,
'global_filter_params': {
            'up_factor': 10.0,
            'down_factor': 5.0,
            'period': 5,
        },
}
```

```python
campaign_equity_filtered, global_filter_data = SwarmFilter.filter_equity(campaign_equity, GLOBAL_FILTER_CONTEXT['global_filter_function'], GLOBAL_FILTER_CONTEXT['global_filter_params'])

title(global_filter_data['name'])
global_filter_data['input_equity'].plot(label='Campaign equity without filter (input)')
global_filter_data['values'].plot(label='Global filter')
campaign_equity_filtered.plot(label='Filtered campaign equity')
legend(loc=2);
```
