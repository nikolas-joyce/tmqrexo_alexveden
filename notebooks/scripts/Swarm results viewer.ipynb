```python
%pylab inline
%load_ext autoreload
%autoreload 2
```

```python
swarms = {}
for file in os.listdir('./swarms/'):
    swarm_name = file.replace('.swm','')
    swarms[swarm_name] = SwarmManager.load(os.path.join('swarms', file))
```

```python
campaign_dict = {}
for swarm_name, smgr in swarms.items():
        smgr = swarms[swarm_name]
        campaign_dict[swarm_name] = smgr.swarm_picked.sum(axis=1)
   
    
campaign_portfolio = pd.DataFrame(campaign_dict)
```

### Campaign members equities

```python
figsize(10,10)
campaign_portfolio.plot(legend=False)
```

### Final Net profit of each swarm

```python
campaign_portfolio.tail(1).T.iloc[:, 0].sort_values(ascending=False)
```
